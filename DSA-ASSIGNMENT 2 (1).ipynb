# Data Visualization

```python

plt.bar(data['Classification'],data['SL'],width=0.5)
plt.ylabel('Sepal Length')
plt.xlabel('Classification')
plt.title('SEPAL LENGTH OF DIFFERENT CLASSES')
plt.show()
```

```python
fig,axs = plt.subplots(2,2,figsize=(8,8))

axs[0,0].bar(data['Classification'],data['SL'],width = 0.5,color='orange')
axs[0,0].set_title('Sepal length of different Species')


axs[0,1].bar(data['Classification'],data['SW'],width=0.5)
axs[0,1].set_title('Sepal width of different Species')


axs[1,0].bar(data['Classification'],data['PL'],width=0.5,color='green')
axs[1,0].set_title('Petal length of different Species')


axs[1,1].bar(data['Classification'],data['PW'],width=0.5,color='red')
axs[1,1].set_title('Petal width of different Species')


plt.tight_layout()

plt.show()
```

From the above plots we can see that

1) Iris-virginica has largest sepal length among the three species and Iris-setosa has the smallest.
2) Iris-setosa has the largest sepal width and Iris-versicolor has the smallest.
3) Iris-Virginica has the largest Petal Length and Iris-setosa has the smallest Petal Length.
4) In case of Petal Width Iris-Virginica has the largest value and Iris-Setosa has the smallest value.

```python
fig,axs = plt.subplots(2,2,figsize=(8,8))


axs[0,0].hist(data['SL'],color='orange',bins=5)
axs[0,0].set_title('Sepal length')


axs[0,1].hist(data['SW'],bins=6)
axs[0,1].set_title('Sepal width')

axs[1,0].hist(data['PL'],color='green',bins=7)
axs[1,0].set_title('Petal length')

axs[1,1].hist(data['PW'],color='red',bins=8)
axs[1,1].set_title('Petal width')

plt.tight_layout()

plt.show()
```

From the above plot we can see that

1) The highest frequency of sepal length is between 40 to 45 which is for sepl length betweem 5.75cm to 6.5 cm
2) The highest frequency of sepal width is around 48 which is for sepal width around 3
3) The highest frequency of petal length is around 48 which is for petal length betwnne 1 to 2
4) The highest frequency of petal width is around 42 which is for petal width between 0.2 to 0.
